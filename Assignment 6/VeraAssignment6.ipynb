```python
#add in a line plot with styles as shown below:
#Line Style dotted and Line-color should be red. Add a circle marker. Line marker color as red Line, the width should be 3.
profitList = df['total_profit']#use tolist() iteratively go through the dataset and set months into a list.
monthList  = df['month_number']#use tolist() iteratively go through the dataset and set months into a list.

profitList.tolist()
monthList.tolist()


#put your plot here.
plt.plot(profitList, 'ro--', markerfacecolor='black', lw='3', label='Profit Data of Last Year')
#X label name = Month Number
plt.xlabel('Month Number')
#Y label name = Sold units number
plt.ylabel('Profit in Dollar ($)')
#Show legend at the lower right location.
plt.legend(loc='lower right')

plt.title('Company Sales data of last year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()


# Explain
#   This plot shows that the company sales increases as the year goes on. 
#   There is three peak months such as July, August, and Novemer
```

### Challenge 4: Read hamburgers and pork product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.
### Explain in detail what this plot displays and the intuition behind it. Discuss the findings of the plot.

```python
# Load the data needed
porkSales = df['pork']
burgerSales = df['hamburgers']

plt.bar([a-0.25 for a in monthList], burgerSales, width= 0.25, label = 'Hamburgers sales data', align='edge')
plt.bar([a+0.25 for a in monthList], porkSales, width= -0.25, label = 'Pork sales data', align='edge')

#label the x with month number, label the y with sales units in number, place the legend in upper left
plt.xlabel('Month Data')
plt.ylabel('Sales Units In Number')
plt.legend(loc='upper left')

#set the xticks with monthList
plt.xticks(monthList)

plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Hamburger and Pork sales data')
plt.show()

#Explain 
#   This graph shows that Pork out sells Hamberger by a fair margin. Expecially in the later months
```

### Challenge 6:  Calculate total sales and plot it in a pie chart.

### Explain in detail what this plot displays and the intuition behind it. Discuss the findings of the plot.

```python
#create a list of labels for each product
labels = ['Hamburgers', 'Brats', 'Pork', 'Steaks', 'Chicken', 'Lamb']
cols = ['hamburgers', 'brats', 'pork', 'steaks', 'chicken', 'lamb']

#create another list called salesdata similar to the beginning of challenge 2. You will want to sum each dataframe in the list.
sales_data = df[cols].sum()

#set the axis text to equal.
plt.title('Pie Sales Data')

#create a pie chart using your salesdata list as input, your labels you created, and an autopct='%1.1f%%'
plt.pie(sales_data, labels=labels, autopct='%1.1f%%')

# Set Legend 
plt.legend(loc='lower right', bbox_to_anchor=(1.6,0))

# Set Title 
plt.title('Pie Sales data')

plt.show()


# Explain 
#   This graph states that products net between 200k and 250k in progit. 
```

```python
plt.plot([],[],color='m', label='Hamburgers', linewidth=5)
plt.plot([],[],color='c', label='Brats', linewidth=5)
plt.plot([],[],color='r', label='Pork', linewidth=5)
plt.plot([],[],color='k', label='Steak', linewidth=5)
plt.plot([],[],color='g', label='Chicken', linewidth=5)
plt.plot([],[],color='y', label='Lamb', linewidth=5)

# Set colors list 
meat_sales_color = ['m', 'c', 'r', 'k', 'g', 'y']

#create a stackplot here. input the monthList, and all category sales data. Set the colors in a list to m, c, r, k, g, y
plt.stackplot(monthList, df['hamburgers'], df['brats'], df['pork'], df['steaks'], df['chicken'], df['lamb'], colors=meat_sales_color)

plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()

# Explain
#   This graph stacks the values on top of each other and shows the total sales as well as the individual sales. This makes it easy to decipher information such as what item is performing well per month 
```
